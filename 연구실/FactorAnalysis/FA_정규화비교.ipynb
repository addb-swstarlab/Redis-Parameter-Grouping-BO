```python
import pandas as pd
from factor_analyzer import FactorAnalyzer
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
df_re = pd.read_csv('bfi.csv', index_col=0)
df_re = df_re.drop(columns=['gender', 'education', 'age'], axis=0)
```

```python
df_re.head()
```

```python
# 정규화
std_scaler = StandardScaler()
df = df_re.copy()
std_scaler.fit(df_re)
df_re = std_scaler.transform(df_re)
df_re = pd.DataFrame(df_re, columns = df.columns, index = list(df.index.values))
```

```python
df_re.head()
```

```python
fa2 = FactorAnalyzer(method='principal', n_factors=3, rotation='varimax')
fa2.fit(df_re)
print(pd.DataFrame(fa2.loadings_, index=df_re.columns))
```
