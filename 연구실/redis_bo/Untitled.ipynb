```python
# -*- coding: utf-8 -*-
import pandas as pd
from knobs import *
from func import *

knobs_DictList = {}

# 리스트로 초기화
for knob in knobs:
    knobs_DictList[knob] = []

config_start = 1
config_end = 2
    
for idx in range(config_start, config_end+1):
    
    # 1.config 파일에서 파라미터 value를 parmas_dict로 복사
    knobs_dict = Extract_knobs("config"+str(idx)+".conf", knobs)

    for knob in knobs_DictList:
        knobs_DictList[knob].append(knobs_dict[knob])


knobs_DataFrame = pd.DataFrame(knobs_DictList)

knobs_DataFrame.to_csv('./CSVfile.csv', index=None) #csv파일로 생성
```
