## 투빅스 과제 중...

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('ggplot') # matplotlib 테마 사용
plt.rc('font', family="Apple SD Gothic Neo") # Mac 컴퓨터에서 지원하는 한글 폰트 사용
plt.rc("axes", unicode_minus=False)
%matplotlib inline

data = pd.read_csv("./example_data.csv", encoding="UTF8")

hist_data = ['temp', 'atemp', 'humidity', 'windspeed', 'casual', 'registered', 'count'] # 히스토그램을 그릴 데이터 산출
```

```python
data[hist_data].hist(figsize=(10,7), bins=30)
plt.suptitle("이상치 제거 전 히스토그램", size='xx-large')
plt.show()
```

```python
sns.pairplot(data[hist_data])
```

```python
data[hist_data].corr()
```

```python
sns.heatmap(data[hist_data].corr(),cmap = 'YlGnBu', annot=True) # annot를 사용해 숫자 표시.
plt.show()
```

분포 형태가 비슷해야 상관계수가 높아지는 듯하다.
