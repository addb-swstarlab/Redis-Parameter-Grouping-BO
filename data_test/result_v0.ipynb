```python
# 평균
def mean_ten_list(data):
    mean_dict = {}
    len_data = len(data)
    count = len_data - 10+1
    for i in range(count):
        print(i, end=' ')
        print(data[i:i+10].mean())
        mean_dict[i] = data[i:i+10]
    return mean_dict
```
